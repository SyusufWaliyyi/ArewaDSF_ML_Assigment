```python
from sklearn.datasets import load_linnerud
from sklearn.linear_model import LinearRegression
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset
df = load_linnerud()
```

```python
# data shape
df.data.shape
```

```python
# target shape
df.target.shape
```

```python
# data feature names
df.feature_names
```

```python
# target feature names
df.target_names
```

```python
# Split data
X = df.data  # Exercise data
y = df.target  # Physiological data
```

```python
# Convert to DataFrame for clarity
X_df = pd.DataFrame(X, columns=df.feature_names)
print(X_df.head())
```

```python
y_df = pd.DataFrame(y, columns=df.target_names)
print(y_df.head())
```

```python
# Split data into features and target
X_jumps = X_df[['Jumps']]
X_situps = X_df[['Situps']]
X_chins = X_df[['Chins']]

y_weight = y_df['Weight']
y_waist = y_df['Waist']
y_pulse = y_df['Pulse']
```

__Waist as a target__

```python
# Model: Waist vs Situps
model = LinearRegression()
model.fit(X_situps, y_waist)

# Predict and plot
y_pred_situps = model.predict(X_situps)
y_pred_situps[:5]
```

```python
# Plot the data
plt.scatter(X_situps, y_waist, color='blue', label='Actual')

# Plot the regression line
plt.plot(X_situps, y_pred_situps, color='red', label='Predicted')
plt.xlabel('Situps')
plt.ylabel('Waist')
plt.title('Waist vs Situps')
plt.legend()
plt.show()
```

```python
# Model: Waist vs Jumps
model = LinearRegression()
model.fit(X_jumps, y_waist)

# Make predictions
y_pred_jumps = model.predict(X_jumps)
y_pred_jumps[:5]
```

```python
# Plot the data
plt.scatter(X_jumps, y_waist, color='blue', label='Actual')

# Plot the regression line
plt.plot(X_jumps, y_pred_jumps, color='red', label='Predicted')
plt.xlabel('Jumps')
plt.ylabel('Waist')
plt.title('Waist vs Jumps')
plt.legend()
plt.show()
```
