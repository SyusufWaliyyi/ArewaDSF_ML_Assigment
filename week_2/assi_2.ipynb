# Filter out the required bird to work on i.e snow geese

```python
# Subsetting the snow goose
Snow_goose = birds[birds['Name']=='Snow goose']
Snow_goose
```

Let plot the lenght and the body mass ato observe the correlation 

```python
x = (Snow_goose['MinLength'], Snow_goose['MaxLength'])
y = (Snow_goose['MinBodyMass'], Snow_goose['MaxBodyMass'])

y1 = (Snow_goose['MinWingspan'], Snow_goose['MaxWingspan'])
```

# Line Plot

```python
plt.plot(x,y,color='red')
plt.title('Length vs Body Mass')
plt.xlabel("Length (cm)")
plt.ylabel("Body Mass (Kg/cm2)")
plt.grid(alpha=0.4)
plt.show()
```

```python
plt.plot(x,y1,color='red')
plt.title('Length vs Wing span')
plt.xlabel("Length (cm)")
plt.ylabel("Wing span (cm)")
plt.grid(alpha=0.4)
plt.show()
```

```python
plt.plot(y1, y,color='red')
plt.title('Body Mass vs Wing Span')
plt.xlabel("Wing Span (cm)")
plt.ylabel("Body Mass (Kg/cm2)")
plt.grid(alpha=0.4)
plt.show()
```

# Scatter Plot

```python
plt.scatter(x,y,color='red')
plt.title('Length vs Body Mass')
plt.xlabel("Length (cm)")
plt.ylabel("Body Mass (Kg/cm2)")
plt.grid(alpha=0.4)
plt.show()
```

```python
plt.scatter(x,y1,color='red')
plt.title('Length vs Wing span')
plt.xlabel("Length (cm)")
plt.ylabel("Wing span (cm)")
plt.grid(alpha=0.4)
plt.show()
```

```python
plt.scatter(y1, y,color='green')
plt.title('Body Mass vs Wing Span')
plt.xlabel("Wing Span (cm)")
plt.ylabel("Body Mass (Kg/cm2)")
plt.grid(alpha=0.4)
plt.show()
```
