## Filter out the required bird to work on i.e snow geese

```python
# Subsetting the snow goose
Snow_goose = birds[birds['Name']=='Snow goose']
Snow_goose
```

## Forming cartesian coordinate to understanding the relationship between the instances

The minimum and the maximum instances were extracted from the row of interest to create the quantities for the various plots.

The minimum and the maximum values signifies the range of values from the minimum value measured to the largest value of its length, body mass and wing span.

Since this is a range, a numpy function linspace was employed to define the possible reading of the columns, which by default create 50 readings. By this its assume 50 samples of snow geese were taken. The number of samples can be adjusted by adding a third argument to the linspace arguments. However, for this the default data points will be use.

```python
# Length range
x = np.linspace(Snow_goose['MinLength'], Snow_goose['MaxLength'])

# Body mass range
y = np.linspace(Snow_goose['MinBodyMass'], Snow_goose['MaxBodyMass'])

# Wing span range
z = np.linspace(Snow_goose['MinWingspan'], Snow_goose['MaxWingspan'])
```

## Line Plot

```python
# Body Mass against Length
plt.plot(x,y,color='red')
plt.title('Length vs Body Mass')
plt.xlabel("Length (cm)")
plt.ylabel("Body Mass (Kg/cm2)")
plt.grid(alpha=0.4)
plt.show()
```

```python
# Wing Span against Length
plt.plot(x,z,color='red')
plt.title('Length vs Wing span')
plt.xlabel("Length (cm)")
plt.ylabel("Wing span (cm)")
plt.grid(alpha=0.4)
plt.show()
```

```python
# Wing Span and Body Mass
plt.plot(z, y,color='red')
plt.title('Body Mass vs Wing Span')
plt.xlabel("Wing Span (cm)")
plt.ylabel("Body Mass (Kg/cm2)")
plt.grid(alpha=0.4)
plt.show()
```

The general conclusion about the snow geese, belonging to the family of __anatide__ of __anser__ genus and order __anserformes__ of ducks/waterfowl category, is that. It's length, mass and wing span are propotional to each other. The larger it's length the heavier it will be and the longer will be it's wing span.
